```python
from agents import enable_verbose_stdout_logging,function_tool

enable_verbose_stdout_logging()
@function_tool
def get_current_weather(city:str)->str:
    """Get the current weather in a given city
    Args:
    city:City name
    Returns:
    Current weather:str
    """
    return f"The weather in {city} is sunny"
agent = Agent(
    name = "Assistan",
    instructions="you are a help agent take the query and answer that",
    model = model,
    tools=[get_current_weather]
)

response = Runner.run_sync(
    starting_agent= agent,
    input = "what is the weather of karachi"

)
print(response.final_output)
```

```python
from agents import enable_verbose_stdout_logging,function_tool

enable_verbose_stdout_logging()
@function_tool
def get_current_news(industry:str)->str:
    """Get the current news for particular industry
    Args:
    industry:The industry to look in current news
    Returns:
    news
    """
    return f"{industry} have new concepts of Agent Native Cloud"
agent = Agent(
    name = "Assistan",
    instructions="you are a help agent take the query and answer that",
    model = model,
    tools=[get_current_weather,get_current_news]
)

response = Runner.run_sync(
    starting_agent= agent,
    input = "what is the weather of karachi and share the latest news in tech?"

)
print(response.final_output)
```
